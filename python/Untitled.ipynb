```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import random

top = Tk()
top.title("test")
top.geometry("100x100")
        
## 함수 설정부

class Game:
    
    win = 0
    draw = 0
    lose = 0
    
    def __init__(self):
        pass
    
    # 가위바위보 매칭 결과 return
    def match(self, a, b):
        if a == "가위" and b == "보":
            self.win += 1
            return "승리!"
        elif a == "바위" and b == "가위":
            self.win += 1
            return "승리!"
        elif a == "보" and b == "바위":
            self.win += 1
            return "승리!"
        elif a == b:
            self.draw += 1
            return "무승부"
        else:
            self.lose += 1
            return "패배 ㅋㅋㅋ"

    # 상대편 가위바위보 선택 결과 return
    def choice_generate(self):
        rsp_list = ["가위", "바위", "보"]
        i = random.randint(0,2)
        return rsp_list[i]

    def print_status(self):
        return str(self.win) + "승 " + str(self.draw) + "무 " + str(self.lose) + "패"

game = Game()

# 버튼 클릭 이벤트 처리
def btnclick():
    your_choice = combo_box.get()
    if your_choice != DEFAULT_VALUE:
        computer_choice = game.choice_generate()
        result = game.match(your_choice, computer_choice)
        msg = messagebox.showinfo(title = "결과", message=result + "\n상대편의 선택 : " + computer_choice + "\n당신의 전적 : " + game.print_status())
    else:
        msg = messagebox.showinfo(title = "말 좀 들어", message="왜 안 고르세요?")

# 콤보박스 설정부
items = ["가위", "바위", "보"]
DEFAULT_VALUE = "고르세요" # 콤보박스 초기값
combo_box = ttk.Combobox(top, height=4, values=items, state="readonly")
combo_box.set(DEFAULT_VALUE)
combo_box.pack()

# 버튼 설정부
btn = Button(top, text="test", padx=10,pady=10,fg="red", command=btnclick)
btn.pack()
top.mainloop()
```
